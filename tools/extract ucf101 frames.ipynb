```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
folders = []
for root, dirs, files in os.walk('./UCF101/'):
    for folder in sorted(dirs):
        folders.append(os.path.join(root,folder))
folders[:3] #take first 3
```

```python
def get_videos_from_folder(folder_path):
    video_list = []
    for root, dirs, files in os.walk(folder_path):
        for name in sorted(files):
            video_list.append(os.path.join(root,name))

    return video_list
```

```python
def extract_video_frames(destination_folder,video_path):
    vid = cv2.VideoCapture(video_path)
    folder_path = video_path.split('/')[3].split('.')[0]
      
    try:
        if not os.path.exists(f'./{destination_folder}/{folder_path}'):
            os.makedirs(f'./{destination_folder}/{folder_path}')
    except OSError:
        print ('Error: Creating directory of data')
        return
      
    currentframe = 0
      
    while(True):
        ret,frame = vid.read()
        if ret:
            name = f'./{destination_folder}/{folder_path}/frame{str(currentframe)}.jpg'
            cv2.imwrite(name, frame)
            currentframe += 1
        else:
            break
```
